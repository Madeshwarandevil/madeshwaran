```python
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
import numpy as np
import pandas as pd
```

```python
norm=pd.DataFrame({'x':np.random.randint(-100,100,5000),
                  'y':np.random.randint(-200,100,5000),
                  'z':np.random.randint(0,1000,5000)})
```

```python
norm
```

```python
ax=plt.axes(projection='3d')
ax.scatter3D(norm.x,norm.y,norm.z)
plt.figure()
```

```python
from sklearn.preprocessing import Normalizer
```

```python
normal=Normalizer()
```

```python
norm=normal.fit_transform(norm)
```

```python
norm
```

```python
norm_mn=pd.DataFrame(norm,columns=['A','B','C'])
```

```python
norm_mn
```

```python
bx=plt.axes(projection='3d')
bx.scatter3D(norm_mn.A,norm_mn.B,norm_mn.C)
plt.figure()
```

# Missing vales

```python
import numpy as np
import pandas as pd
```

```python
data={'I score':[54,np.nan,54,np.nan],
     'II score':[65,76,87,np.nan],
     'III score':[82,43,np.nan,np.nan]}
```

```python
score=pd.DataFrame(data)
```

```python
score
```

```python
score.isnull()
```

```python
score.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
```

```python
imputer.fit(score)
```

```python
imputer.transform(score)
```

```python
imputermedian=SimpleImputer(missing_values=np.nan,strategy='median')
```

```python
imputermedian.fit(score)
```

```python
imputermedian.transform(score)
```
